#### apply()=> Apply is a pandas function which will apply specified functionality to each value of series

```python
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
```

```python
df['Last_Name']
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace("[^0-9]", "",regex=True)
```

```python
l1 = lambda x:x[0:3]+"-"+x[3:6]+"-"+x[6:10]
```

```python
df['Phone_Number']=df['Phone_Number'].apply(l1)
```
